# 그림 파일 다운로드

- 그림 파일 다운로드에는 
    1. 그림파일의 주소
    2. 그림 파일의 다운로드 지정
    이 필요합니다.
    
- 이를 처리하기 위해 그림파일 주소를 가져오는 크롤러 하나와
- 그 그림 파일을 다운 받는 크롤러 하나가 필요합니다.

```python
#라이브러리 임포트
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import urllib.request
```

```python
# 크롬 드라이버 세팅
driver = webdriver.Chrome("chromedriver.exe")

# 그림 파일 url 저장을 위한 빈 리스트
imgs = []

# 1 ~ 757화까지 자동으로 순회하는 크롤러(1회 딜레이 추가)
for i in range(1, 101):    
    driver.get("https://sports.donga.com/cartoon?cid=0100000204&sid=%s" % i)
    time.sleep(1)

    # 소스코드 가져오기
    source = driver.page_source #v
    
    # 파싱 하기
    s1 = BeautifulSoup(source, "html.parser")
    
    # div 태그의 cartoon_view 클래스 요소 가져오기
    cartoon = s1.find("div", class_="cartoon_view") # 한 개가 확실하면 find(인덱스 번호 안 써도 됨) 써도 됨
    #cartoon = s1.find("div", class_="cartoon_view").find("img") # find만 가능
    # 가져온 자료 내부의 img 태그 가져오기
    c_img = cartoon.find("img")

    #img 태그 내부의 "src" 속성 내용물 가져오기
    c_src = c_img.get("src") 
    #가져온 src 속성(다운로드 링크주소) 리스트에 집어 넣기
    imgs.append(str(c_src))
    print(c_src)
```

# 수집해온 경로의 그림 파일 다운로드

- urllib.request.urlretrieve(다운받을 주소, 파일 경로 및 이름)

```python
imgs
```

```python
count = 0
for down_url in imgs:
    count += 1
    # 다운받을 url, 다운로드할 퐁더 및 파일 이름을 순차적으로 넣으면 다운된다, 
    urllib.request.urlretrieve(down_url, "C:/Users/user/data0102/20200201/cartoon/%s화.jpg" % count)
    print("%s화 다운로드 완료" % count)
    time.sleep(0.1)
```
