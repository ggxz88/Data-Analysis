```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import codecs
```

```python
driver = webdriver.Chrome("chromedriver.exe") #v

for i in range(251):

    driver.get("http://www.yes24.com/24/category/bestseller?CategoryNumber=001&sumgb=06&fetchSize=40&PageNumber=%s" % i)
    
    source = driver.page_source #v
    
    s1 = BeautifulSoup(source, "html.parser")#v
    
    book_list = s1.find_all("td", class_="goodsTxtInfo")
    
    time.sleep(2)
    rank = 0
    for book in book_list: #v
        time.sleep(2)
        rank += 1
        print("%d위 도서 정보" % rank)
        best_list = book.find_all(["p", "div"])
        #print(best_list[0].text.strip())
        print(best_list[0].find_all("a")[0].text.strip())#오류
        print(best_list[1].text.replace("\n", ""))
        print(best_list[2].text)
        print("----------------------------------------------------------------------")

        #나, 틀림
        #print(best_list)
        #print(book_list[book].find_all("p").text)
        #print(book_list.find_all("div", class_ = "aupu"))
        #print(book_list[book].find_all("p").text)



```
