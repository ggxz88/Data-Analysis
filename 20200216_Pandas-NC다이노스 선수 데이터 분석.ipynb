## 5. 평균 안타수 기준 상위 10위 타자들의 3년간 안타, 홈런, 타율 조회

#### 참고) NaN 값이 있을 경우 '-'로 대체해서 표현

```python
# 1. 전체 데이터 중에서 보고싶은 값을 기준으로 구조 변경 => 해당 선수 데이터만 추출
tmp = nc_result1.set_index(['연도', '선수명'])[['안타', '홈런', '타율']].unstack(0)
tmp
# NaN을 다른 값으로 채울 때는 .fillna(채움값)을 사용
tmp = tmp.fillna('-') # 일시적으로 바뀐 결과만 보여주기 때문에 저장해야 함
tmp
```

```python
# 2. 전체 데이터 -> 보고싶은 값 -> 구조 변경 -> 해당 선수들의 데이터만 추출
# 보고 싶은 값 : 평균 안타수 기준 상위 10위 타자들의 3년간 '안타', '홈런', '타율'
# 추출 방법 : 인덱싱 / 메서드
tmp2 = nc_result1[['선수명', '연도', '타율', '홈런', '안타']].set_index(['연도'])
tmp2
```

```python
# 안타개수 상위 10위 선수들 목록을 뽑아주세요 v
#hit10 = tmp2.sort_values(by='안타', ascending=False).head(10)
hit10 = tmp2.pivot_table(index='선수명', values='안타', aggfunc='sum').sort_values(by='안타', ascending=False).head(10)
```

```python
hit10 = hit10.head(10).index.tolist()
hit10 
```

```python
# 특정 데이터프레임에서 내가 제시한 명단에 해당하는 로우만 남기기
tmp3 = tmp2[tmp2.선수명.isin(hit10)]
tmp3
```

```python
# 연도, 선수명을 로우로 반환
# 로우에 있던 연도를 컬럼으로 교환
tmp3 = tmp3.reset_index()
# 연도, 선수명으로 로우 설정
tmp3 = tmp3.set_index(['연도', '선수명'])
# unstack()을 이용해 연도를 컬럼으로 이용
# fillna('-')로 NaN 처리
tmp3 = tmp3.unstack(0).fillna('-')
tmp3
```
