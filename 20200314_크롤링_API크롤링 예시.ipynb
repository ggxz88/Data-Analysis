```python
# 모든 데이터를 append로 쌓아서 상점에 대한 정보는 store_info 변수에
# 데이터프레임 형태로 저장하고
# 재고에 대한 정보는 sale_info 변수에 데이터프레임 형태로 저장한 뒤
# 이 둘을 합친 데이터를 total_info 데이터에 저장
import pandas as pd
store_json = []
sale_json = []
```

```python
for i in range(1, 7):
    time.sleep(1)
    data = urllib.request.urlopen('https://8oi9s0nnth.apigw.ntruss.com/corona19-masks/v1/stores/json?page=%s&perPage=5000' % i)
    store_json.append(data)
```

```python
store_list = []
for data in store_json:
    json_data = data.read().decode('utf-8')
    json_data = json.loads(json_data)
    store_list.append(json_normalize(json_data['storeInfos']))
len(store_list)
```

```python
store_data = store_list[0]
store_data
```

```python
for df in store_list[1:]:
    store_data = store_data.append(df, ignore_index=True)
```

```python
store_data
```

```python
for i in range(1, 6):
    time.sleep(1)
    data = urllib.request.urlopen('https://8oi9s0nnth.apigw.ntruss.com/corona19-masks/v1/sales/json?page=%s&perPage=5000' % i) 
    sale_json.append(data)
```

```python
sale_list = []
for data in sale_json:
    json_data = data.read().decode('utf-8')
    json_data = json.loads(json_data)
    sale_list.append(json_normalize(json_data['sales']))
len(sale_list)
```

```python
sale_data = sale_list[0]
sale_data
```

```python
for df in sale_list[1:]:
    sale_data = sale_data.append(df, ignore_index=True)
```

```python
total_data = pd.merge(store_data, sale_data)
```

```python
total_data
```
