```python
# 라이브러리 임포트
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import codecs
```

```python
# 크롬 드라이버 파일 세팅
driver = webdriver.Chrome("chromedriver.exe")

# 알라딘 베스트셀러 페이지로 이동
driver.get("https://www.aladin.co.kr/shop/common/wbest.aspx?BranchType=1&start=we")

# 소스코드 가져오기
source = driver.page_source
print(source)
```

```python
# 소스코드 파싱하기
s1 = BeautifulSoup(source,"html.parser")

#len(s1.find_all("li")) #730
book_list = s1.find_all("div", class_="ss_book_box") #50
```

```python
len(book_list) # 50
type(book_list) #list 자료형 : 인덱스로 하나하나 꺼내야 함
```

```python
# 위의 코드를 50번 반복해서 1위~50위 책의 제목, 저자, 가격순으로 출력해주세요
#나
for idx in range(len(book_list)):
    print("%d위" % (idx + 1))
    for li in book_list[idx].find_all("li")[1:4]:
        print(li.text)
    print("-----------------------------")
```

```python
for book in book_list:
    li_list = book.find_all("li")
    if len(li_list) == 7:
        for li in li_list[0:3]:
            print(li.text)
    else:
        for li in li_list[1:4]:
            print(li.text)
    print("-------------------")
```

```python
# 텍스트 파일로(result2.txt)에 순위별로 출력
a_rank = codecs.open("C:/Users/user/data0102/20200119/result2.txt", encoding = "utf-8", mode="w" )

idx = 0
for book in book_list:
    li_list = book.find_all("li")
    idx += 1
    a_rank.write("%d위 서적 정보\r\n" % idx)
    if len(li_list) == 7:
        for li in li_list[0:3]:
            a_rank.write(li.text+"\r\n")
    else:
        for li in li_list[1:4]:
            a_rank.write(li.text+"\r\n")
    a_rank.write("-------------------\r\n")
    
a_rank.close()
```
