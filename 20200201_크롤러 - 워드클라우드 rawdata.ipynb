### 워드클라우드 제작에 필요한 raw data를 가져오는 크롤러 만들기

- 워드클라우드는 텍스트 자료에서 언급된 단어를 빈출순서대로 시각화해주는 방법
- 따라서 먼저 텍스트자료가 필요하다.
- 크롤링을 통해 워드클라우드에 사용할 텍스트 자료를 가져올 수 있다.

```python
#라이브러리 임포트
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import codecs
```

```python
#다음 사회 기사
driver = webdriver.Chrome("chromedriver.exe")

link_list = [] #v

for i in range(1, 20):
    driver.get("https://news.daum.net/breakingnews/society/affair?page=%s&regDate=20200131" % i) #v
    
    time.sleep(2)
    
    source = driver.page_source #v
    
    s1 = BeautifulSoup(source, "html.parser") #v
        
    contents = s1.find_all("div", class_="box_etc")
    
    for content in range(len(contents)):
        titles = contents[content].find_all("a", class_="link_txt")

        for title in range(len(titles)):
            link_list.append((str(titles[title]).split('">')[0].split('href="')[1].replace(";","&")))
    print(link_list)
  

driver.close()
```

```python
driver = webdriver.Chrome("chromedriver.exe")
d_news = codecs.open("C:/Users/user/data0102/20200201/result_daum.txt", encoding="utf-8", mode="w")

for link in link_list:
    time.sleep(2)
    driver.get(link)
    
    source = driver.page_source
    
    s1 = BeautifulSoup(source, "html.parser")
    
    articles = s1.find_all("div", id="harmonyContainer", class_="article_view")
    
    d_news.write(articles[0].text)
    d_news.write("-----------------------------------------------------------")
    
d_news.close()
```

#### 링크 수집 네이버 세계 - 아시아 호주

```python
driver = webdriver.Chrome("chromedriver.exe")

# url 저장을 위한 빈 리스트
urls = [] #v

for i in range(1, 21):
    # 사용자가 입력한 날짜 페이지로 이동
    driver.get("https://news.naver.com/main/list.nhn?mode=LS2D&mid=shm&sid2=231&sid1=104&date=20200131&page=%s" % i)
    
    time.sleep(2)
    
    #소스코드 가져오기
    source = driver.page_source
    
    #파싱
    s1 = BeautifulSoup(source, "html.parser")
    
    s_div =  s1.find_all("div", class_="list_body newsflash_body")
    
    s_alist = s_div[0].find_all("a")
    
    #urls 내부에 링크 주소 다 담기(단 중복된 요소 존재)
    for a_tag in s_alist:
        #a 태그 내부의 href 속성에 있는 값만 가져오기
        href = a_tag.get("href")
        urls.append(str(href)) #동일한 주소 두 개씩 출력
        
print(len(urls))

#urls 정보 수집 완료
driver.close()
```

```python
# urls 내부의 중복자료 없애기
# 중복 자료를 없애기 위해서는 리스크를 집합 자료형으로 바꿨다가 리스트로 되돌립니다.
print(len(urls))
# urls 내부의 중복자료를 전부 1개만 남기고 소거해 저장
# 집합 자료형은 내부 중복 요소가 존재하는 것을 허용하지 않습니다.
urls = set(urls)
# 집합 자료형을 다시 리스트 자료형으로 되돌리기
urls = list(urls)
# 줄어든 갯수 확인
print(len(urls))
```

```python
# 기사 페이지 별로 접근시키는 크롤러 만들기
# 크롬 드라이버 세팅
driver = webdriver.Chrome("chromedriver.exe")
delete_content = """// flash 오류를 우회하기 위한 함수 추가
function _flash_removeCallback() {}"""
n_news = codecs.open("C:/Users/user/data0102/20200201/result_naver.txt", encoding = "utf-8", mode = "w")
#기사 388개의 본문 페이지에 순차적으로 접근
for url in urls:
    time.sleep(1)
    driver.get(url)
    
    # 기사 전체 소스코드 가져오기
    source = driver.page_source
        
    # 파싱하기
    s1 = BeautifulSoup(source, "html.parser")
   
    # 본문에서 기사 부분만 find_all로 가져오기
    #articles = s1.find_all("div", class_="article_body _font_setting_target size3 font1")
    # .find()는 가장 먼저 접근한 한 개 만 가져옴, 보통 단일 요소를 가져올 때 사용, 한 개 뿐이라서 리스트 아님
    # .find()로 가져오는 자료는 리스트가 아니기때문에 인덱싱할 필요가 없음 
    article = s1.find("div", id="articleBodyContents", class_="_article_body_contents") 
    
    # id와 class가 다른 기사가 있어서 에러가 남
    # try - except 구문은 try 내부의 코드 실행 중 만약 모종의 에러가 발생한다면
    # except 내부의 코드로 대체해 진행하라는 명령어 입니다.
    # 방법 1: 다르면 건너 뛰기
    #try:
    #    n_news.write(article.text.replace(delete_content, "")+"\r\n")
    #except:
    #    pass
    # 방법 2
    try:
        n_news.write(article.text.replace(delete_content, "")+"\r\n")
    except:
        article = s1.find("div", id="newsEndContents", class_="news_end font1 size3") 
        n_news.write(article.text.replace(delete_content, "")+"\r\n")

    n_news.write("--------------------------\r\n")
    
n_news.close()
```
