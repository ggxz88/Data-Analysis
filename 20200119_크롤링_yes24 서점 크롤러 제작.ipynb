```python
# 빈 리스트를 만들고 거기다가 num1~num40까지 채워넣는 반복문을 작성하세요
# 단 19, 20 은 _line이 추가로 붙음
class_list =[]

for i in range(40):
    if (i == 18) or (i == 19):
        class_list.append("num%s_line" % (i+1)) # %d가 x
    else :
        class_list.append("num%s" % (i+1))
        
print(class_list)
```

```python
# 순위별로 책 정보를 출력하는 코드를 작성해주세요
# 40개의 li 내부에서 제목, 저자, 가격 순으로 자료를 뽑아서 나열하기
book_list = s1.find_all("li", class_=class_list)
len(book_list)# 40
idx = 0
for book in book_list: #v
    idx += 1
    print("%d위 서적 정보" % idx)
    p_list = book.find_all("p")
    for p in p_list[2:5]:
        print(p.text.strip())
    print("---------------------------------")
```

```python
# 콘솔창에 출력한 해당 자료를 result3.txt 파일에 출력해주세요
y_rank = codecs.open("C:/Users/user/data0102/20200119/result3.txt", encoding = "utf-8", mode = "w")

idx = 0
for book in book_list:
    p_list = book.find_all("p")
    idx += 1
    y_rank.write("%d위 서적 정보\r\n" % idx)
    for p in p_list[2:5]:
        y_rank.write(p.text+"\r\n")
    y_rank.write("---------------------------------\r\n")

y_rank.close()
```
